```python
import pandas as pd
import os
os.chdir('D:\Project\Twitter_depression_detector\data')
```

```python
def split_dataframe(df, chunk_size = 3000): 
    chunks = list()
    num_chunks = len(df) // chunk_size + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunk_size:(i+1)*chunk_size])
    return chunks
```

```python
df=pd.read_csv('cleaned_merged_depression_tweets.csv')
```

```python
df=df.drop(['mentionedUserCount'],axis=1)
```

```python
chunks=split_dataframe(df)
```
