# Get positive tweets from sent140 dataset

```python
import os
import pandas as pd
os.chdir(r'D:\Project\Twitter_depression_detector\data')
```

```python
df=pd.read_csv("sentiment140.csv",names=['target','ids','date','flag','user','renderedContent'])
```

```python
df.columns
```

```python
df=df[df.target==4]
print(len(df))
```

```python
df=df.sample(n=95842,random_state=1)
```

```python
df
```

```python
df
```

```python
df['mentioned_users'] = df.renderedContent.str.findall(r'(?<![@\w])@(\w{1,25})').apply(','.join)
df['mentioned_users']=[string.split(",") if len(string)!=0 else [] for string in df['mentioned_users']]
df
```

```python
len(df.iloc[3][-2])
```

```python
len(df.iloc[1][-2])
```

```python
df['mentionedUserCount']=[len(mentioned_user) for mentioned_user in df['mentioned_users']]
```

```python
df
```

```python
df=df[['renderedContent','mentionedUserCount']]
df.to_csv("positive_tweets.csv",index=False) 
```
