```python
import pandas as pd
import glob

path = r'D:\Project\Twitter_depression_detector\data\tweet_trend_dataset\10_days_dataset' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame.to_csv("Depression_merged.csv")
```
